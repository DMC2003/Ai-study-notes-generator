```python
import os
from groq import Groq
from IPython.display import display, Markdown

# 1. Setup - PASTE YOUR KEY INSIDE THE QUOTES
GROQ_API_KEY = "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"
client = Groq(api_key=GROQ_API_KEY)

def generate_study_notes(topic, complexity):
    prompt = f"""
    You are an expert academic professor. Create highly structured, detailed study notes on the topic: {topic}.
    The notes should be for a {complexity} level of understanding. Make it interactive with emojis

    Please include:
    1. **Introduction**: A brief overview.
    2. **Key Concepts**: 5-7 important terms explained in details.
    3. **Detailed Explanation**: A deep dive into the logic/mechanics along with formulas if needed.
    4. **Examples**: Real-world applications.
    5. **Quiz**: 3 practice questions with answers hidden at the bottom.

    Format everything beautifully using Markdown headers and bullet points.
    """

    # Using Llama 3.3 70B - a world-class AI model
    completion = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
    )
    return completion.choices[0].message.content

# 2. Form Interface
# @title 📝 AI Study Notes Generator
Topic_name = "Machine Learning" # @param {type:"string"}
Level = "Undergraduate" # @param ["High School", "Undergraduate", "PhD Researcher"]

if GROQ_API_KEY == "PASTE_YOUR_GROQ_KEY_HERE":
    print("❌ Error: Please paste your Groq API key in the code!")
else:
    try:
        print(f"🚀 AI is thinking... generating notes for {topic_name}")
        notes = generate_study_notes(topic_name, level)
        print("-" * 30)
        display(Markdown(notes))
    except Exception as e:
        print(f"❌ An error occurred: {e}")
```
